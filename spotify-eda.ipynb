### Graphs using Seaborn and Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### 1.Songs distribution across various languages in spotify.

```python
languages= df["language"].value_counts()
sl = pd.DataFrame(languages).reset_index()
sl
```

```python
plt.figure(figsize=(10,5))
plt.pie(sl["count"], labels=sl["language"], autopct='%1.1f%%',pctdistance=0.9,labeldistance=1.2)
plt.legend(title="Languages",loc="upper left", bbox_to_anchor=(1, 1))
plt.title("Songs Across Different Languages")
plt.show()
```

#### **Insights** : English songs are the highest in numbers in spotify and malayalam songs are the least. About 13000 songs are of unknown languages.

### 2.Top artists in English and the number of songs they released.

```python
Eng_artists = df[df["language"]=="English"].groupby(["artist_name"])["track_name"].count().reset_index()
top_eng = Eng_artists.sort_values("track_name",ascending=False).head(10)
top_eng= top_eng.rename(columns={"track_name": "song_count"})
top_eng
```

```python

sns.barplot(x=top_eng["artist_name"],y=top_eng["song_count"],color="lightgreen")
plt.xticks(rotation=45) 
plt.title("Top Artists in English")
```

### 3. Find the artist and language of the most popular song in 2024.

```python
latest = df[df["year"]==2024].sort_values(by="popularity")
top_song = latest[["track_name","popularity","language"]].tail(1)
top_song
```

```python
# Top 50 songs of 2024.
top_50 =latest[["track_name","popularity","language"]].tail(50)
sns.scatterplot(y=top_50["popularity"],x=top_50["language"])
plt.xticks(rotation=45)
plt.title("Top 50 songs by languages")
```

#### Most popular song is an English song. Songs with popularity greater than 80 are in English and Korean languages.

### 4. Find the songs of Shreya Ghoshal and how many songs did she sing in each year.

```python
Shreya_songs =df[df["artist_name"]=="Shreya Ghoshal"]
Shreya_songs_year = Shreya_songs["year"].value_counts().reset_index()
Shreya_songs_year
```

```python
sns.barplot(x="year",y="count",data=Shreya_songs_year)
plt.xticks(rotation =90)
plt.yticks(np.arange(0,130,10))
plt.title("Songs sang in each year by Shreya Ghoshal")
```

#### Shreya Ghoshal sang highest number of songs in the year 2018, then in 2019 and least in 2002. Recently she has been singing less songs compared to past years.

### 5. Total number of songs released in last 15 years. 

```python
last_15 = df["year"].value_counts().head(15).reset_index()
last_15
```

```python
sns.lineplot(x=last_15["year"],y=last_15["count"],marker ="p",color ="maroon")
plt.xticks(np.arange(2010,2025,1),rotation = 45)
plt.yticks(np.arange(0,7000,1000))
plt.xlabel("Year")
plt.ylabel("Number of songs")
plt.title("Songs Released last 15 years")
```

#### Number of songs is highest in the year 2023 and lowest in year 2010. More songs are released in the year 2013 compared to that of from 2014 to 2020.
